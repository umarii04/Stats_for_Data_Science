```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = titanic_df['age'].quantile(0.25)
Q3 = titanic_df['age'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
titanic_df = titanic_df[(titanic_df['age'] >= lower_bound) & (titanic_df['age'] <= upper_bound)]
titanic_df.reset_index(drop=True, inplace=True)

# Display the updated DataFrame
titanic_df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic_df, x="age" , hue='sex')
```

```python
# Generate data for normal, uniform, and exponential distributions
normal_data = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution
uniform_data = np.random.uniform(low=0, high=1, size=1000)  # Uniform distribution
exponential_data = np.random.exponential(scale=1, size=1000)  # Exponential distribution

# Plot the distributions
plt.figure(figsize=(15, 5))

# Normal distribution
plt.subplot(1, 3, 1)
plt.hist(normal_data, bins=30, color='blue', alpha=0.7)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Uniform distribution
plt.subplot(1, 3, 2)
plt.hist(uniform_data, bins=30, color='green', alpha=0.7)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Exponential distribution
plt.subplot(1, 3, 3)
plt.hist(exponential_data, bins=30, color='red', alpha=0.7)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
